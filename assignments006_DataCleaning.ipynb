```python
import pandas as pd
import numpy as np
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN','londON_StockhOlm','Budapest_PaRis', 'Brussels_londOn'],
                   'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
                   'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
                   'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )','12. Air France', '"Swiss Air"']})
df
```

```python
df.dtypes
```

```python
df1=df.copy()
```

```python
df1['FlightNumber']=df.FlightNumber.interpolate().astype(int)
df1
```

```python
df1.dtypes
```

```python
def From_split(x):
    l=str(x).split("_")
    return l[0]
```

```python
df1['From']= df1.From_To.apply(From_split)
df1
```

```python
def To_split(x):
    l=str(x).split("_")
    return l[1]
```

```python
df1['To']= df1.From_To.apply(To_split)
df1
```

```python
df1['From'] = df1.From.str.title()
df1["To"] = df1.To.str.title()
df1
```

```python
df1.drop(['From_To'], axis = 1, inplace=True)
df1
```

```python
df_delays = pd.DataFrame(df1.RecentDelays)
df_delays
```

```python
df_delays.RecentDelays.values.tolist()
```

```python
df_delays.index
```

```python
# creates a colum for each element of the nested lists

df2_delays = pd.DataFrame(df_delays.RecentDelays.values.tolist(), index=df_delays.index)
df2_delays
```

```python
df2_delays = df2_delays.rename(columns=lambda x:f'Delay_{x+1}')
df2_delays
```

```python
df3 = df1.join(df2_delays)
df3
```

```python
df3.drop(['RecentDelays'],inplace=True, axis=1)
df3
```
